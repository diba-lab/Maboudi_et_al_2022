```python
import matplotlib.pyplot as plt

colors = plt.cm.tab20(range(20))
plt.imshow([colors], aspect = 'auto')
plt.axis('off')
plt.show()
print(colors)
```

```python
import matplotlib.pyplot as plt

colors = plt.cm.Set2(range(17))
plt.imshow([colors], aspect = 'auto')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt

colors = plt.cm.Dark2(range(10))
plt.imshow([colors], aspect = 'auto')
plt.axis('off')
plt.show()
```

```python
import numpy as np
from matplotlib.cm import viridis, cividis
from scipy.io import savemat

cmap = cividis(np.linspace(0, 1, 256))

savemat('cividis.mat', {'cmap':cmap})
```
